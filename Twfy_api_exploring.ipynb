The above code shows that 'speaker' and 'parent' are the keys to other dictionaries --> flatten them to get easy access to their content

```python
#This function flattens the dictionaries in ['rows'] 
def flatten_dict(d, parent_key='', sep='_'):   
    items = [] #list to store the flattened key-value pair
    for k, v in d.items():
        new_key = f"{parent_key}{sep}{k}" if parent_key else k #Create a new key by adding the current key to the parent key
        if isinstance(v, dict): #If the value is a dictionary recursively flatten it
            items.extend(flatten_dict(v, new_key, sep=sep).items())
        elif isinstance(v, list): #if the value is a list iterate over the list
            for i, item in enumerate(v):
                new_item_key = f"{new_key}{sep}{i}" # Create a new key for each item in the list, appending the index
                if isinstance(item, dict): # If the list item is a dictionary, recursively flatten it
                    items.extend(flatten_dict(item, new_item_key, sep=sep).items())
                else:
                    items.append((new_item_key, item)) # Otherwise, add the item directly to the list
        else:
            items.append((new_key, v)) # If the value is neither a dict nor a list, add it directly
    return dict(items)
```

```python
list_flat_getData = []
for item in explore_rows:
    flatten = flatten_dict(item)
    list_flat_getData.append(flatten)

# Create DataFrame
df_flat_data = pd.DataFrame(list_flat_getData)
df_flat_data
```
