Load the model

```python
from joblib import load

model_path = "/Users/tom.willcocks/Documents/code/discovery_utils/tests/ml_experimentation/heat_pump_model_logreg.joblib"

# Load the model from the file
clf = load(model_path)
```

Load and prepare the data

```python
# Define base name
base_name = "organization_descriptions"

data_path = f"/Users/tom.willcocks/Downloads/{base_name}.csv"
```

```python
if base_name == 'organization_descriptions':
    col = "description"
elif base_name == 'organizations':
    col = "short_description"
```

```python
import pandas as pd

# Load the CSV file into a DataFrame
df = pd.read_csv(data_path)

# Select only the specified columns
df = df[['id', 'name', col]]
```

Choose how you want to filter your data for testing

```python
# # OPTION 1: Sample a random batch of 100 entries from the DataFrame
# sampled_df = df.sample(n=100, random_state=40)  # Adjust the random_state for reproducibility

# OPTION 2: Assuming df is your DataFrame and 'col' is the name of the column to check
sampled_df = df[df[col].str.contains("heat pump", case=False, na=False)]

# Reset the index of the sampled DataFrame
sampled_df = sampled_df.reset_index(drop=True)
```

Apply model

```python
from sentence_transformers import SentenceTransformer
model = SentenceTransformer("all-MiniLM-L6-v2")

# Replace this with your actual embeddings generation process
sampled_embeddings = model.encode(sampled_df['description'])
```

```python
# Assuming 'sampled_embeddings' are your input features for the sampled data
predictions = clf.predict(sampled_embeddings)
```

Print results

```python
import numpy as np

# Count occurrences
ones_count = np.sum(predictions == 1)
zeros_count = np.sum(predictions == 0)

print(f"Number of 1s: {ones_count}")
print(f"Number of 0s: {zeros_count}")
print("\n\n\n")

for i in range(len(sampled_df)):
    print(f"Name: {sampled_df['name'][i]}")
    print(f"Description: {sampled_df[col][i]}")
    print(f"Prediction: {predictions[i]}")
    print("\n")
```
